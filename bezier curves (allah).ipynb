```python
import math
from matplotlib.pyplot import *
def combinaison_lineaire(A,B,u,v):
    return [A[0]*u+B[0]*v,A[1]*u+B[1]*v]
def interpolation_lineaire(A,B,t):
    return combinaison_lineaire(A,B,t,1-t)
def point_bezier_3(points_control,t):
    x=(1-t)**2
    y=t*t
    A = combinaison_lineaire(points_control[0],points_control[1],(1-t)*x,3*t*x)
    B = combinaison_lineaire(points_control[2],points_control[3],3*y*(1-t),y*t)
    return [A[0]+B[0],A[1]+B[1]]
def courbe_bezier_3(points_control,N):
    if len(points_control) != 4:
        raise SystemExit("4 points de controle")
    dt = 1.0/N
    t = dt
    points_courbe = [points_control[0]]
    while t < 1.0:
        points_courbe.append(point_bezier_3(points_control,t))
        t += dt
    points_courbe.append(points_control[3])
    return points_courbe
def plot_points(points_courbe,style='-'):
    x = []
    y = []
    for p in points_courbe:
        x.append(p[0])
        y.append(p[1])
    plot(x,y,style)

############allah#################
figure(figsize=(13.5,13.5))
P0 = [12,8]
P1 = [14,-6]
P2 = [10,-8]
P3 = [8.3,-7]
points = courbe_bezier_3([P0,P1,P2,P3],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P4 = [8,-7]
P5 = [7.5,-5.5]
P6 = [7.5,-5.5]
points = courbe_bezier_3([P3,P4,P5,P6],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P4,P5,P6],style='o')

P7 = [7.5,-5.5]
P8 = [6,-10]
P9 = [1.6,-7.8]
P10 = [0.5,-4]
points = courbe_bezier_3([P7,P8,P9,P10],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P4,P5,P6],style='o')

P10 = [0.5,-4]
P11 = [-0.5,2.5]
P12 = [-1,-8]
P13 = [-2.5,-9]
points = courbe_bezier_3([P10,P11,P12,P13],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P4,P5,P6],style='o')


P17 = [12,8]
P18 = [10,6]
P19 = [10,6]
P20 = [12,8]
points = courbe_bezier_3([P17,P18,P19,P20],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P21 = [10.5,6.3]
P22= [12,-6]
P23 = [10,-6]
P24 = [9,-5]
points = courbe_bezier_3([P21,P22,P23,P24],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P24 = [9,-5]
P25 = [8,-3]
P26= [8,6]
P27 = [9,8]
points = courbe_bezier_3([P24,P25,P26,P27],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P27 = [9,8]
P28 = [6,6]
P29 = [6,6]
P30= [9,8]
points = courbe_bezier_3([P27,P28,P29,P30],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P30= [6.8,6.5]
P31 = [7,-4.5]
P32 = [6.4,-4.5]
P33 = [5.5,-5.4]
points = courbe_bezier_3([P30,P31,P32,P33],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P33 = [5.5,-5.4]
P34= [4,-6.9]
P35 = [1,-7]
P36 = [0.5,0]
points = courbe_bezier_3([P33,P34,P35,P36],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P36 = [0.5,0]
P37 = [-1,2.5]
P38= [-1.5,-2]
P39 = [-2.3,-6]
points = courbe_bezier_3([P36,P37,P38,P39],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P361 = [-2.3,-6]
P371 = [-2.7,-7.5]
P381= [-5,-9]
P391 = [-6,-0]
points = courbe_bezier_3([P361,P371,P381,P391],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P39 = [-6,-0]
P40 = [-6,-8]
P41 = [-4,-10]
P42= [-2.5,-9]
points = courbe_bezier_3([P39,P40,P41,P42],50)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')




##########alifon########


P0 = [4,-16]
P1 = [4,-7]
P2 = [4,7]
P3 = [4,16]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')


P0 = [4,-16]
P1 = [5.3,-7]
P2 = [5.3,7]
P3 = [5.3,14]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')



P0 = [4,16]
P1 = [3.5,16.5]
P2 = [3,17]
P3 = [2.5,17.5]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [2.5,17.5]
P1 = [2.9,18.3]
P2 = [3.3,18.9]
P3 = [4,19.5]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [4,19.5]
P1 = [5,17.7]
P2 = [6,16]
P3 = [7.5,13]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [7.5,13]
P1 = [6.8,13.3]
P2 = [5,13.7]
P3 = [5.3,14]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

###########symbole haut##############


P0 = [5.5,18.7]
P1 = [7.3,16.8]
P2 = [8,15.4]
P3 = [7.3,14]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [5.9,20]
P1 = [7.5,17.2]
P2 = [8.1,14.7]
P3 = [7.3,14]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [5.5,18.7]
P1 = [5.6,18.8]
P2 = [5.67,19.7]
P3 = [5.7,20]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

##############chadda############
   

P0 = [9.5,12.3]
P1 = [10,11.3]
P2 = [10.5,9.7]
P3 = [9,9.75]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')


P0 = [9.7,12.6]
P1 = [9.5,12.3]
P2 = [9.5,12.3]
P3 = [9.5,12.3]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')


P0 = [9.7,12.6]
P1 = [10.3,11]
P2 = [10.8,8.9]
P3 = [8,9.5]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [8.1,12.45]
P1 = [8.1,12.45]
P2 = [8.1,12.45]
P3 = [7.9,12.3]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [5.7,12.6]
P1 = [5.4,11]
P2 = [5,8.9]
P3 = [8,9.5]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [5.7,12.6]
P1 = [5.7,12.6]
P2 = [5.7,12.6]
P3 = [5.9,12.45]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [5.9,12.45]
P1 = [5.9,11.3]
P2 = [5.3,9.7]
P3 = [6.8,9.75]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [8.1,12.45]
P1 = [8.1,11.3]
P2 = [7.5,9.7]
P3 = [9,9.75]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')


P0 = [7.9,12.3]
P1 = [7.8,11.3]
P2 = [8.3,9.7]
P3 = [6.8,9.75]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

####################h#######################

P0 = [-0.3,3]
P1 = [-3,2.7]
P2 = [-1.5,3]
P3 = [-2,3.5]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')



P0 = [-0.4,3]
P1 = [-3,3]
P2 = [-1.5,3]
P3 = [-2.8,3.4]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')


P0 = [-2.8,3.4]
P1 = [-3,7.8]
P2 = [-1.5,6.3]
P3 = [-1.4,3.2]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [-1.4,3.2]
P1 = [-2,2.1]
P2 = [-2.1,1.8]
P3 = [-6,1]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [-2.8,3.4]
P1 = [-3,7]
P2 = [-1.5,5.5]
P3 = [-1.4,3.2]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [-1.4,3.2]
P1 = [-1.9,2.7]
P2 = [-2.1,2.5]
P3 = [-6,1]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')


P0 = [-0.4,3]
P1 = [-3,4]
P2 = [-1.5,3.3]
P3 = [-2.8,4.1]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [-3,7.8]
P1 = [-1.1,5.8]
P2 = [0,4.9]
P3 = [-0.2,2.7]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')


P0 = [-0.2,3.5]
P1 = [-3,7.6]
P2 = [-3,7.4]
P3 = [-3,6.5]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')


###########symboles############


P0 = [9,-8]
P1 = [9.4,-12.3]
P2 = [9.7,-12.3]
P3 = [10,-10.2]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [9,-8]
P1 = [9.1,-13.2]
P2 = [10.1,-13.2]
P3 = [10,-10.2]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [-1.1,-4.5]
P1 = [-0.7,-8.8]
P2 = [-0.5,-8.8]
P3 = [-0.1,-6.7]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')

P0 = [-1.1,-4.5]
P1 = [-1,-9.7]
P2 = [-0,-9.7]
P3 = [-0.1,-6.7]
points = courbe_bezier_3([P0,P1,P2,P3],100)
plot_points(points,style='b.')
plot_points(points,style='r-')
plot_points([P0,P1,P2,P3],style='o')
```
